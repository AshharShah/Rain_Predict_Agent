# Rain Fall Prediction
A notebook that creates a machine learning model to predict if it will rain tomorrow based on the data we have recieved from a .csv file with different climate features.

```python
# importing our libraries that are to be used
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# setting the path where our csv file is located that contains the data set
path = "./Dataset/weatherAUS.csv"
```

```python
# importing the csv as a dataframe to clean it
df = pd.read_csv(path)
df
```

```python
# removing all the null/none/NaN values from our coloumns by removing the associated row
df = df.dropna(axis=0)
```

```python
df
```

```python
# drop the date column as it is useless
df = df.drop(["Date"], axis=1)
df
```

```python
# dropping the location as it has a wide variety of values which results in useless features and may lower model performance
df = df.drop(["Location"], axis=1)
df
```

```python
# dropping the wind gust direction as it does not matter
df = df.drop(["WindGustDir"], axis=1)
df
```

```python
df = df.drop(["WindDir9am"], axis=1)
df
```

```python
df = df.drop(["WindDir3pm"], axis=1)
df
```

```python
df = df.drop(["WindSpeed9am"], axis=1)
df = df.drop(["WindSpeed3pm"], axis=1)
df
```

```python
df = df.drop(["Pressure9am"], axis=1)
df = df.drop(["Pressure3pm"], axis=1)
df
```

```python
# mapping all Yes values to 1 and all No values to 0
df['RainToday'] = df['RainToday'].map({'Yes': 1, 'No': 0})
df['RainTomorrow'] = df['RainTomorrow'].map({'Yes': 1, 'No': 0})
```

```python
df
```

```python
# X.shape should be (N, M) where M >= 1
X = df[['MinTemp', 'MaxTemp', 'Rainfall','Evaporation', 'Sunshine', 'WindGustSpeed', 'Humidity9am', 'Humidity3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm', 'RainToday']]
# y.shape should be (N, 1)
y = df['RainTomorrow']
```

```python
X.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
logisticRegr = LogisticRegression(solver='lbfgs', max_iter=400)
```

```python
logisticRegr.fit(X_train, y_train)
```

```python
predictions = logisticRegr.predict(X_test)
```

```python
score = logisticRegr.score(X_test, y_test)
print(str(score*100) + "%")
```
