# Rain Fall Prediction
A notebook that creates a machine learning model to predict if it will rain tomorrow based on the data we have recieved from a .csv file with different climate features.

```python
# importing our libraries that are to be used
import pandas as pd
import numpy as np
```

```python
# setting the path where our csv file is located that contains the data set
path = "./Dataset/weatherAUS.csv"
```

```python
# importing the csv as a dataframe to clean it
df = pd.read_csv(path)
df
```

```python
# removing all the null/none/NaN values from our coloumns by removing the associated row
df = df.dropna(axis=0)
```

```python
df
```

```python
df = df.drop(["Date"], axis=1)
df
```

```python
df = df.drop(["Location"], axis=1)
df
```

```python
df = df.drop(["WindGustDir"], axis=1)
df
```

```python
df = df.drop(["WindDir9am"], axis=1)
df
```

```python
df = df.drop(["WindDir3pm"], axis=1)
df
```

```python
df = df.drop(["WindSpeed9am"], axis=1)
df = df.drop(["WindSpeed3pm"], axis=1)
df
```

```python
df = df.drop(["Pressure9am"], axis=1)
df = df.drop(["Pressure3pm"], axis=1)
df
```

```python
df['RainToday'] = df['RainToday'].map({'Yes': 1, 'No': 0})
df['RainTomorrow'] = df['RainTomorrow'].map({'Yes': 1, 'No': 0})
```
